# **XGBoost**

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('/kaggle/input/breast-cancer2benign/Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

**The Dependant Variable is not in the standard binary form(0,1)**

```python
print(y)
```

##Lable encoding the Dependant Variable

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
print(y)
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Feature Extraction using PCA, LDA, Kernel-PCA

**Feature Scaling before Feature Extraction**

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```

**Feature Extraction**

Using Principal component analysis
 (PCA)

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 1)
xg_X_train_PCA = pca.fit_transform(X_train_scaled)
xg_X_test_PCA = pca.transform(X_test_scaled)
```

Using Linear discriminant analysis (LDA)

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda = LDA(n_components = 1)
xg_X_train_LDA = lda.fit_transform(X_train_scaled,y_train)
xg_X_test_LDA = lda.transform(X_test_scaled)
```

Using Kernel-Principal component analysis
 (K-PCA)

```python
from sklearn.decomposition import KernelPCA
kpca = KernelPCA(n_components = 1, kernel = 'rbf')
xg_X_train_kpca = kpca.fit_transform(X_train_scaled)
xg_X_test_kpca = kpca.transform(X_test_scaled)
```

## Training XGBoost on the Training set

```python
from xgboost import XGBClassifier
classifier1 = XGBClassifier()
classifier2 = XGBClassifier()
classifier3 = XGBClassifier()
classifier4 = XGBClassifier()
classifier5 = XGBClassifier()
classifier1.fit(X_train, y_train)
classifier2.fit(X_train_scaled, y_train)
classifier3.fit(xg_X_train_PCA, y_train)
classifier4.fit(xg_X_train_LDA, y_train)
classifier5.fit(xg_X_train_kpca, y_train)
```

## Making the Confusion Matrix for different cases

```python
from sklearn.metrics import confusion_matrix, accuracy_score
print('For normal Features : ')
y_pred = classifier1.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(accuracy_score(y_test, y_pred))
print('For scaled Features : ')
y_pred = classifier2.predict(X_test_scaled)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(accuracy_score(y_test, y_pred))
print('For Features Extracted via PCA : ')
y_pred = classifier3.predict(xg_X_test_PCA)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(accuracy_score(y_test, y_pred))
print('For Features Extracted via LDA : ')
y_pred = classifier4.predict(xg_X_test_LDA)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(accuracy_score(y_test, y_pred))
print('For Features Extracted via K-PCA : ')
y_pred = classifier5.predict(xg_X_test_kpca)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(accuracy_score(y_test, y_pred))
```

## Applying k-Fold Cross Validation for different cases

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier1, X = X_train, y = y_train, cv = 10)
print('For normal Features : ')
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
accuracies = cross_val_score(estimator = classifier2, X = X_train_scaled, y = y_train, cv = 10)
print('For scaled Features : ')
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
accuracies = cross_val_score(estimator = classifier3, X = xg_X_train_PCA, y = y_train, cv = 10)
print('For Features Extracted via PCA : ')
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
accuracies = cross_val_score(estimator = classifier4, X = xg_X_train_LDA, y = y_train, cv = 10)
print('For Features Extracted via LDA : ')
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
accuracies = cross_val_score(estimator = classifier5, X = xg_X_train_kpca, y = y_train, cv = 10)
print('For Features Extracted via K-PCA : ')
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

Analyzing the results from XGBoost and CatBoost across different feature engineering techniques provides valuable insights:

   - Both models exhibit similar performance with minimal deviation.

2. **Scaled Features:**
   - **XGBoost:** 96.71%, CatBoost: 97.26%
   - Scaled features maintain the same performance as normal features for both models.
   - Feature scaling does not significantly impact model accuracy in this case.

3. **Features Extracted via PCA:**
   - **XGBoost:** 96.88%, CatBoost: 97.06%
   - Both models achieve slightly improved accuracy compared to normal features.
   - PCA demonstrates a positive impact on model performance.

4. **Features Extracted via LDA:**

   - Both models show enhanced accuracy compared to the original and scaled features.
   - Linear Discriminant Analysis (LDA) is effective in capturing relevant information.

5. **Features Extracted via K-PCA:**
   - **XGBoost:** 94.87%, CatBoost: 95.78%
   - Both models experience a decrease in accuracy compared to other feature sets.
   - Kernel PCA introduces non-linearity that may not benefit the models in this context.

**Conclusion:**
- CatBoost consistently demonstrates slightly higher accuracy than XGBoost across all feature sets.
- Feature scaling has a minimal impact on both models.
- PCA and LDA contribute positively to model accuracy.
- Kernel PCA may not be as effective in this context, resulting in a decrease in accuracy.

The choice between XGBoost and CatBoost depends on the specific requirements of the problem and the significance of computational efficiency, as CatBoost generally performs well "out of the box" without extensive tuning.
